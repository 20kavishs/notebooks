## *Vaccine efficacy*

### a. 

We use our knowledge of conditional probability to calculate below. 

$$P(\text{severe|vax}) = \frac{\text{num. severe and vax}}{\text{num. vax}} = \frac{290+11}{3501117+2133515} = 5.34 \cdot 10^{-5}
$$

$$P(\text{severe|unvax}) = \frac{\text{num. severe and unvax}}{\text{num. unvax}} = \frac{43+171}{116833+186077}
= 1.64 \cdot 10^{-4}$$

### b.

The measure $\frac{P(\text{severe|vax)}}{P(\text{severe|unvax})}$ describes the ratio of severe cases given a vaccination and severe cases given a non-vaccination. 

Subtracting this term from 1 provides the ratio of non-severe cases given a vaccination to non-severe cases given no vaccination, or $\frac{P(\text{nonsevere|vax)}}{P(\text{nonsevere|unvax})}$. This is our measure of efficacy. 

Consider an efficacy of 0.9. We are conveying the chance of being non-severe if you are vaccinated is 90% of the chance of being non-severe if you are unvaccinated (or, expressed differently, that the chance of being severe if you are vaccinated is 10% the chance of being severe if unvaccinated).  Efficacy shows the degree to which getting the vaccine affects your chances of becoming a severe case as compared to no vaccine, which is a more informative criterion for the desired question.  

$$1-\frac{P(\text{severe|vax})}{P(\text{severe|unvax})} = 0.674$$


### c.

The quantity given provides a ratio of your chance of being vaccinated and severe to your chance of being unvaccinated and severe. This tells an individual they are less likely to be severe if they are severe and vaccinated, but does not compare their chances before and after being vaccinated (i.e. it does not inform them how much a vaccine will affect their chances of illness). 


### d. 

We again use the rules of conditional probability for our computation. 

$$P_1 = P(\text{severe|vax, over 50}) = \frac{\text{num. severe and vax and over 50}}{\text{num. vax over 50}} = \frac{290}{2133515} = 1.36 \cdot 10^{-4}
$$

$$P_2 = P(\text{severe|unvax, over 50}) = \frac{\text{num. severe and unvax and over 50}}{\text{num. unvax over 50}} = \frac{171}{186077}
= 9.19 \cdot 10^{-4}$$

$$P_3 = P(\text{severe|vax, under 50}) = \frac{\text{num. severe and vax and under 50}}{\text{num. vax under 50}} = \frac{11}{3501117} = 3.14 \cdot 10^{-6}
$$

$$P_4 = P(\text{severe|unvax, under 50}) = \frac{\text{num. severe and unvax and under 50}}{\text{num. unvax under 50}} = \frac{43}{116833}
= 3.85 \cdot 10^{-5}$$



### e. 

Efficacy for over 50 = $$1-\frac{P_1}{P_2} = 0.852$$

Efficacy under 50 = $$1-\frac{P_3}{P_4} = 0.918$$

The vaccine is more effective for those under 50, and this may for several reasons. For one, it apppears the group of individuals over 50 has a higher proportion of severe unvaccinated cases and severe vaccinated cases. Instead of being a vaccine problem, this could involve the increased susceptibility of older individuals to the illness, which would change our measurement of efficacy comparatively because a higher proportion of unvaccinated older people will get severe symptoms. 


### f. 

We must compute 4 probabilities. We observe the total number of participants under 50 is $$1116833+43+3501117+11 = 4618004$$ 

$$P(\text{vaxxed AND not-severe|under 50}) = \frac{3501117-11}{4618004} = 0.758$$
$$P(\text{unvaxxed AND not-severe|under 50}) = \frac{1116833-43}{4618004} = 0.242$$
$$P(\text{vaxxed AND severe|under 50}) = \frac{11}{4618004} = 2.382 \cdot 10^{-6}$$
$$P(\text{unvaxxed AND severe|under 50}) = \frac{43}{4618004} = 9.311 \cdot 10^{-6}$$


We scale the first two probabilities by 1000 to create boxes (factor D1 below). We scale the second two probabilities by $10^7$ to create boxes (factor D2 below). 

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import math


P_vax_nonsevere = 0.758
P_unvax_nonsevere = 0.242

#scale factor 1
D1 = 1000

P_vax_severe = 2.382 * 10 ** (-6)
P_unvax_severe = 9.311 * 10 ** (-6)

#Scale factor 2 - just to make these boxes visible it is greater than D1
D2 = 10**(7)

#make rectangles with scaled probabilities
box_vax_nonsevere = patches.Rectangle((0,0), math.sqrt(D1*P_vax_nonsevere),  math.sqrt(D1*P_vax_nonsevere), 0,facecolor='blue')
box_unvax_nonsevere = patches.Rectangle((0,0), math.sqrt(D1*P_unvax_nonsevere),  math.sqrt(D1*P_unvax_nonsevere),  -90 , facecolor='red')

box_vax_severe = patches.Rectangle((0,0), math.sqrt(D2*P_vax_severe),  math.sqrt(D2*P_vax_severe), -180,facecolor='green')
box_unvax_severe = patches.Rectangle((0,0), math.sqrt(D2*P_unvax_severe),  math.sqrt(D2*P_unvax_severe),  -270 , facecolor='orange')


#add rectangles to plot
fig, ax = plt.subplots() 
ax.add_patch(box_vax_nonsevere)
ax.add_patch(box_unvax_nonsevere)
ax.add_patch(box_vax_severe)
ax.add_patch(box_unvax_severe)
ax.set_aspect('equal', adjustable='box')

plt.xlim([-30, 30])
plt.ylim([-30, 30])
ax.title.set_text('Vaccine Efficacy for Individuals under 50')

plt.show()
```

$P(\text{unvaxxed AND not-severe|under 50})$ respectively. We see the chance of being vaccinated and not-severe when under 50 is significantly greater than being unvaccinated and not-severe under 50. 

The two boxes on the top left and bottom left are $P(\text{unvaxxed AND severe|under 50})$ and 
$P(\text{vaxxed AND severe|under 50})$ respectively. We see the chance of being unvaccinated and severe when under 50 is significantly greater than being vaccinated and severe under 50. 

This picture connects to (c) by visually describing the change in the likelihood of getting a severe or nonsevere case for those under 50 between if they are already vaccinated vs. if they are not. ## *Poisson limit*

### a. 
We start with $M=100$ and $\epsilon = 0.03$. We create a function to graph the bimomial distribution with its parameters, and graph the log of this function with the given parameters. We also create a function to graph the Poisson distribution given its parameters (note we use three parameters for this function, and compute $\mu = M\epsilon$).

Then,we graph the logs of the binomial and Poisson distributions together, and likewise for the raw distributions. 

```python
import numpy as np
import matplotlib.pyplot as plt
import math

#Get all values of l
x = np.arange(0, 101, 1)

#binomial distribution
def bin(l, M, epsilon): 
    successes = epsilon**(l)
    failures = (1-epsilon)**(M-l)
    return math.comb(M, l) * successes * failures

#poisson distribution
def poisson(l, M, epsilon): 
    mu=M*epsilon
    return 1/(math.factorial(l)) * (mu**l) * math.exp(-1*mu)

#log of binomial and log of poisson
y = [math.log(bin(i, 100, 0.03)) for i in x]
y2 = [math.log(poisson(i, 100, 0.03)) for i in x]

#plotting
plt.figure(0)

plt.title('Distribution comparisons')

plt.xlabel("l")
plt.ylabel("log(P)")

plt.plot(x, y, label="binomial")
plt.plot(x, y2, label="poisson")
plt.legend(loc="upper left")


#new range to better see graph overlap

x = np.arange(0, 10, 1)

#just raw binomial and poisson
y3 = [bin(i, 100, 0.03) for i in x]
y4 = [poisson(i, 100, 0.03) for i in x]

#plotting
plt.figure(1)
plt.plot(x, y3, label="binomial")
plt.plot(x, y4, label="poisson")
plt.xlabel("l")
plt.ylabel("P")
plt.legend(loc="upper left")
plt.show()
```

### d.

In our derivation, we assumed $\mu = \epsilon M$. Since $\mu$ is a constant, we see $\epsilon \propto \frac{1}{M}$. As $M$ increases, $\epsilon$ approaches 0. We then take the limit as $M$ approaches $\infty$ of the binomial distribution and find the Poisson distribution. As $M$ approaches $\infty$, $\epsilon$ decreases as mentioned, and our Poisson distribution should more closely approximate the binomial distribution. We observe this in our graphs: the Poisson and Binomial distributions fit over each other more closely for smaller $\epsilon$ values.

### a. 

The form of this relation tells use there is a linear relationship between change in expectation and change in variance. Namely, this means that larger mean conductance changes lead to larger fluctuations in conductance, and the two are releated by a constant of proportionality, which we use to find the single channel conductance as Katz and Miledi did. 

To be more specific, we take the parameters from the physical model. We have $M$ channels, $\epsilon$ probability of opening, $l$ channels open at any given time.

After observing the membrane potential $g$ experimentally, we can seek to find a relationship between $<g>$ and $var(g)$.

From our physical model (as computed in a previous homework), we know $<l> = M \epsilon$ and $var(l) = M \epsilon(1-\epsilon)$. We also see that $g=al$, because the excess potential is $al$ in the system.

Then, $<g> = a <l>$ because $a$ is a constant. We see then with $var(l)$ above that $var(g) = a \cdot var(l) = a^2 M \epsilon(1-\epsilon)$. We know $a=var(a)$ because a is a constant. 


We are looking to make $var(g)$ as a function of $<g>$, to work in accordance with the given graph. We see the quantity $\epsilon$ depends on $<g>$: $<g> = a \cdot <l> = a M \epsilon \implies \epsilon = \frac{<g>}{M a}$.

Plugging this in, $var(g) = a^2 M \frac{<g>}{M a}(1-\frac{<g>}{M a}) = a<g>(1-\frac{<g>}{M a})$. As $M$ approaches $\infty$ (we are using a Poisson distribution, and can derive the Poisson from the binomial by taking the limit as M approaches $\infty$), $\frac{<g>}{M a}$ approaches 0, so we have $var(g)=a<g>$, which makes sense as a linear fit of the graph with $a$, the single channel conductance, as the slope. 



### b. 


To solve for $<l>$, we see $var(g) = a <g> \implies var(g) = a^2 <l>$ because $<g> = a \cdot <l>$. From part (a) we are given $var(g)=a<g>$. Thus, $<l> = \frac{var(g)}{a^2} = \frac{a<g>}{a^2} = \frac{(<g>)^2}{var(g)}$.

Subbing in the highest point on the graph, with $<g> \approx 4 \cdot 10^{-6}$ and $var(g) \approx 1 \cdot 10^{-16}$,
we get $<l> = \frac{(4 \cdot 10^{-6})^2}{1 \cdot 10^{-16}} \approx 160000$ channels open. 


We see that at the highest point, ~160,000 channels are open.
